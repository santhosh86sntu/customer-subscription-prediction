# Target data set

## Input 1( Minits watched on Platform)

```python
# loading Student Learning dataset 

print(stlern.shape)
stlern.head()
```

```python
# Droping unwanted columns
# Aggregating minites watched column using groupby function.
# student_id by default set to Index.

std_watch = stlern.drop(columns=['course_id','date_watched'])
std_watch = std_watch.groupby(by=std_watch.student_id, axis=0).sum()
std_watch.head()
```

## Input 2 (Days Engaged with platform) 

```python
# Loading Student Engagement Dataset

print(stengmnt.shape)
stengmnt.head()
```

```python
# Creating duplicate column with 1 up to length of index.

stengmnt['days_engage'] = int(1)
stengmnt.head()
```

```python
# Droping unwanted columns

stengg_count = stengmnt.drop(columns=['engagement_id',
                                      'engagement_quizzes',
                                      'engagement_exams',
                                      'engagement_lessons',
                                      'date_engaged'])
stengg_count.head()
```

```python
# Changing datatype of column "days_engage"
# Aggregating days engage of indivisual students.
# By default setting student_id to index

stengg_count = stengg_count.groupby(by=stengg_count['student_id'], axis=0).sum()
stengg_count.head()
```

## Output ( 1: purchase, 0: free)

# Final Dataset for Modeling

```python
# using union operator to combing data sets.

input1 = std_watch
input2 = stengg_count
input3 = eng_quiz
input4 = eng_exam
input5 = eng_hub_qa
output = st_purch
```

```python
# Loading Student information dataset

print(stinfo.shape)
stinfo.head()
```

```python
# Removing unwanted columns from student information dataset 

df = stinfo.drop(columns=['date_registered','student_country'])
df.head()
```

```python
# Merging od all datasets with respect to student id
# using merge of outer which combines union data.

df = df.merge(input1, how='outer' ,on='student_id')
df = df.merge(input2, how='outer', on='student_id')
df = df.merge(input3, how='outer', on='student_id')
df = df.merge(input4, how='outer', on='student_id')
df = df.merge(input5, how='outer', on='student_id')
df = df.merge(output, how='outer', on='student_id')
```

```python
# Checking of merged values

df.head(10)
```

```python
# Filling null values with "0"

df = df.fillna(0)
```

```python
# Checking null value replace with 0

df.head(10)
```

```python
# Information of df before typecasting

df.info()
```

```python
# Typecasting each column with required format for analysis

df['student_id'] = df['student_id'].astype(int)
df['student_id'] = df['student_id'].astype(str)
df['days_engage'] = df['days_engage'].astype(int)
df['eng_quiz'] = df['eng_quiz'].astype(int)
df['eng_exam'] = df['eng_exam'].astype(int)
df['eng_hub_qa'] = df['eng_hub_qa'].astype(int)
df['st_purch'] = df['st_purch'].astype(int)
```
